# Simple Neural Network By Keras
In this file I am gonig to Implement a simple network by using Keras.
The first step is define a function to load reguired data and its classes.

We use Numpy librari to load the data in it and the Keras library to define the Model.

In this project I wana load the Pima indians onest of diabetes dataset. It descrbes the paitent medical record data wheather they had onset of diabets within the five years. 



The data set links: <br> 
&emsp;&emsp; o [Dataset CSV File (pima-indians-diabetes.csv)](https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv) <br>
&emsp;&emsp; o [Dataset Details](https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.names) <br>
    

    


```python
from numpy import loadtxt
from keras import Sequential
from keras.layers import Dense 

dataset= loadtxt('./pima-indians-diabetes.data.txt' , delimiter=',')
print(type(dataset) ,"The data shape is a matrix with:{} Dimention".format(dataset.shape))

print(dataset[0:2,:])

#split data 
X= dataset[:,:-1]  # features 
y=dataset[:,-1]    # Targets  0 or 1

```

# Define the Model in the Keras:

First we need to create a sequential model in the Keras then we can add as many layers we need to that.
Fully connected layers are defined as a Dense layer in the Keras, and we can determine the number of nodes and activation function for this layer.

```python
# define the model

model= Sequential()
# the first hidden layer has 12 nodes with 8 inputs for each
model.add(Dense(12, input_dim=8,activation='relu'))
# the second hidden layer has 12 inputs and 8 outputs (nodes)
model.add(Dense(8,activation='relu'))
# the output layer has one node
model.add(Dense(1,activation='sigmoid'))
```

# Comiling the Model
Now, we need to compile the cmodel. It is an efficient library which is called backend. backend fide the best way to present your network on the hardware.

Trraining the network means find the best set of the weights to present the relationship between the input and output.
to training the network we need to specify: <br>
-loss function --> cross entropy<br>
-optimizer --> stochastic gradient descent (adam)<br>
-metric<br> 


```python
model.compile( loss= 'binary_crossentropy' ,
                optimizer= 'adam' ,
              metrics=['accuracy'])
```

# Fit the Model

```python
# fit the keras model on the dataset
model.fit(X, y, epochs=150, batch_size=10)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# make probability predictions with the model
predictions = model.predict(X)
# round predictions 
rounded = [round(x[0]) for x in predictions]

# summarize the first 5 cases
for i in range(5):
	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```
